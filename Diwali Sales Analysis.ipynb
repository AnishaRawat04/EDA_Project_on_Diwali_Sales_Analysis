## ***Gender***

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)

# Females have made more purchases than men and below is for how much amount they have made purchases
```

## ***AGE***


```python
# which age group is doing more purchases(we have already concluded from above that it is women section who is doing more purchses , bu this time age group?)

ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
  ax.bar_label(bars)
  
```

***CONCLUSION:***  <span style="font-size:18px;"> 26-35 years of age group people speciallly 3269 thousands females and 1272 no of males have made more purchases </span>


```python
# total sales by each age-group

sales_age_group = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
ax = sns.barplot(x='Age Group', y = 'Amount', data = sales_age_group )

for bars in ax.containers:
    ax.bar_label(bars)
```

## ***State*** 

```python
df.columns
```

```python
#  top 10 states where maximum number of orders are there?

top_states =  df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})  # size of plot is set
ax = sns.barplot(x='State', y='Orders', data=top_states)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  top 10 states where maximum number of total_sales/revenue has been generated?

top_states =  df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})  # size of plot is set
ax = sns.barplot(x='State', y='Amount', data=top_states)

# bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

## most of the orders we have received is from UP, 8th most order from - kerala, haryana and then gujarat
## highest sales is from UP again, but the 8th most highest order is from haryana, then bihar and gujarat

```python
# Marital Status---count--married or unmarried-->who made more purchases:-

ax= sns.countplot(x = 'Marital_Status',data = df)
sns.set(rc={'figure.figsize':(7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.info()
```

```python
# Marital Status---sum--married or unmarried-->total_purchases or total_revenue made by them:-

total_sales_marital_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})  # size of plot is set
ax = sns.barplot(x='Marital_Status', y='Amount', data = total_sales_marital_status, hue='Gender')

# bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

### most of the purchaser are married women from UP, with total_sales/revenue contributed by them is 4+ million

## Occupation

```python
# Occupation---count of purchases--what occupation people r from -->who have made more purchases:-

ax= sns.countplot(x = 'Occupation',data = df)  #, hue='Gender'
sns.set(rc={'figure.figsize':(15,5)})

# data_labels in bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Occupation---sum---which occupation have made more total_sales/revenue generated is more:-

total_sales_occupation = df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})  # size of plot is set
ax = sns.barplot(x='Occupation', y='Amount', data = total_sales_occupation, hue='Gender')

# bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

### IT sector occupation--->people have made more purchases, then healthcare, aviation, and banking
### women in every sector are making highest purchases.

# Product_Category (wise Sales Analysis)

```python
# which product category is purchased more by people? 

ax=sns.countplot(x = 'Product_Category', data= df)
sns.set(rc={'figure.figsize':(35,5)})

# data_labels in bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# which product category has made the more sales?

total_sales_product_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,5)})  # size of plot is set
ax = sns.barplot(x='Product_Category', y='Amount', data = total_sales_product_category)

# bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

### clothing-->food-->electronics-->footwear---> this is most highest purchased product category by people in this order
### food sales: 3+ million sales/revenue---->clothing: 1+ million sales/revenue----->electronics: 1.5+million

## Top selling Products

```python
# top 10 selling product-->which has placed order multiple times---.the product which has been sold more.

top_selling_product= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(45,5)})  # size of plot is set
ax = sns.barplot(x='Product_ID', y='Orders', data = top_selling_product)     #, hue='Gender'

# Add value labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Top 10 Selling Products")
plt.show()
```

```python
# top 10 products -->which has  generated highest total_sales
total_sales_product= df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(45,5)})  # size of plot is set
ax = sns.barplot(x='Product_ID', y='Amount', data = total_sales_product)     #, hue='Gender'

# Add value labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Top 10 highest total sales products")
plt.show()
```

## ***FINAL_CONCLUSION:*** Married women of age group 26-35 years from UP,Maharashtra, and Karnataka working in IT Healthcare and Aviationare more likely to buy products from food, clothing and electronics category.
